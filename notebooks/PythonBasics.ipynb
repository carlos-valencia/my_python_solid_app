## Challenge
### Goal
Format a string value

### Details
Given the username:
```
raw_username = " Py_Thon-ROCKS123!!! "
```

Do the following
1. Remove whitespace
2. Convert all to lowercase
3. Replace '-' with '_'
4. Remove all '!'
5. Check at least one digit is present
6. Check it contains the word 'python'
7. Create a variable that stores only the first 8 characters
8. Print a formatted string (f-string) that displays: the cleaned username, whether it contains a digit, whether it contains 'python', & the first 8 characters


````python
raw_username = " Py_Thon-ROCKS123!!! "

clean_username = raw_username.strip()
clean_username = clean_username.replace("-", "_")
clean_username = clean_username.replace("!", "")
sub_username = clean_username[:8]

if any(char.isdigit() for char in clean_username):
    has_digit = True
else:
    has_digit = False
    
if "python" in clean_username: # will only return true if it's exactly python
    has_python = True
else:
    has_python = False


print(f"Clean Username: {clean_username}\nHas digit?: {has_digit}\nHas Python?: {has_python}" + 
      f"\nFirst 8 characters: {sub_username}")
````

### Sorting Lists

```

thislist.sort(reverse = True) # descending
thislist.sort(key = str.lower) # performs case insensitive sort, no key will be case sensitive

def word_length(word):
    return len(word)

thislist.sort(key = word_length) # sorts list by word length

## Challenge
### Goal
Manipulate Tuples

### Details
Given the following list of tuples, complete the following using only list comprehensions
1. Print only students with an 'A'
2. Print all names (but nothing else)
3. Give all the 20-year old students a 'B' and then print the list

```
students = [
    ("Alice", 20, "A"),
    ("Bob", 19, "B"),
    ("Charlie", 21, "A"),
    ("Diana", 20, "C")
]

````python
students = [
    ("Alice", 20, "A"),
    ("Bob", 19, "B"),
    ("Charlie", 21, "A"),
    ("Diana", 20, "C")
]

print([x for x in students if x[2] == "A"])
print([y[0] for y in students])

students = [
    (x[0], x[1], "B") if x[1] == 20 else (x[0], x[1], x[2])
    for x in students
]

print(students)
````

Manipulate Sets

### Details
Given the following list, complete the following
1. Create a set from the list to remove duplicates
2. Print the total number of attendees
3. Print 'Alice is going!' if Alice is found in the set
4. Add 'Grace' to the set
5. Remove 'Bob' from the set twice (don't throw errors)
6. Print the final complete set of attendees

```
attendees = [
    "Alice", "Bob", "Alice", "Ethan", "Bob", "Diana", "Ethan", "Frank"

## OOP
### Classes
The 'self' parameter must be included as the first parameter in any class method
Self has access to other class methods/variables
We can delete instantiations of a class
We can use 'pass' as a placeholder

### __init__ - totally not a constructor
Under the hood, a `__new__` method gets called to create the object
`__init__` runs after, and initializes object state
The `__str__` method controls what is returned when the object is printed

### Inheritance
- Inherit all methods and properties from another class
- Parent/Base class
- Child/Derived class
```
class Person:
# definition

class Student(Person):
# definition
```

We need to call an `__init__` in our child class so it doesn't generally inherit the one from the parent class
We should use .super() instead of calling the parent directly, so as to not break multiple inheritance

### Function Polymorphism
- Polymorphism refers to functions/classes behaving differently depending on input or object type
- Another example of function polymorphism is len()

### Encapsulation
```__``` before a member variable enforces private access of variable
```_``` before a member variable denotes a protected property, but only a convention, not enforced

### Abstraction
- ABC stands for abstract base class

```
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def move():
        pass

    @property
    @abstractmethod
    def id(self):
        pass

class Car(Vehicle):
    def move(self):
        print("Vroom")

class Plane(Vehicle):
    def move(self):
        print("Weeeee")

    @property
    def id(self):
        return '2'
```
